```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA

%matplotlib inline
```

```python
X1, Y1 = make_blobs(n_features=10, n_samples=150, centers=4, random_state=4,cluster_std=2)
print(X1.shape)
print(X1[0])
```

```python
pca = PCA(n_components=4)
pc = pca.fit_transform(X1)
```

```python
pc_df = pd.DataFrame(data=pc, columns=['C1', 'C2', 'C3', 'C4'])
pc_df['Cluster'] = Y1
pc_df.head()
```

```python
print(pca.explained_variance_ratio_)
```

```python
df = pd.DataFrame({'var': pca.explained_variance_ratio_, 'PC': ['C1', 'C2', 'C3', 'C4']})
sns.barplot(x='PC', y='var', data=df, color='c')
```

```python
sns.lmplot(x='C1', y='C4', data=pc_df, fit_reg=False, hue='Cluster', legend=True, scatter_kws={'s': 80})
```
